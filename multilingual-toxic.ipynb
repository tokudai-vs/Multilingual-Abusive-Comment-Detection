```python
train = pd.read_csv("/kaggle/input/multilingualabusivecomment/ShareChat-IndoML-Datathon-NSFW-CommentChallenge_Train.csv", error_bad_lines=False, warn_bad_lines=False)
```

```python
train.info(verbose=True, null_counts=True)
```

```python
toxic = train[train['label'] == 1]
```

```python
toxic
```

```python
toxic_text = toxic['commentText']
```

```python
toxic_text
```

```python
toxic_words = list(toxic_text)
```

```python
toxic_words
```

```python
words = [word for line in toxic_words for word in line.split()]
```

```python
import collections
counts = collections.Counter(words)
```

```python
counts
```

```python
 first = counts.most_common(1000)
```

```python
first
```

```python
vals=[]

for i in range(len(first)):
    vals.append(first[i][0])
```

```python
vals
```

```python
test = pd.read_csv("/kaggle/input/multilingualabusivecomment/ShareChat-IndoML-Datathon-NSFW-CommentChallenge_Test_20_Percent_NoLabel.csv", error_bad_lines=False, warn_bad_lines=False)
```

```python
test
```

```python
test_text = list(test['commentText'])
```

```python
test_text
```

```python
preds = []

for s in test_text:
    if len([each for each in vals if each.lower() in str(s).lower()])>0:
        preds.append(1)
    else:
        preds.append(0)
```

```python
preds
```

```python
submission = pd.DataFrame({'CommentId': test.CommentId, "Label": preds})
```

```python
submission.to_csv('submission.csv', index=False)
```
